```python
import wikipediaapi
import os

# Initialize the Wikipedia API
wiki = wikipediaapi.Wikipedia(
    language='en',
    user_agent='CricketRAGBot/1.0'
)

# Articles to scrape
cricket_topics = [
    # Basics & Rules
    "Cricket",
    "Laws of Cricket",
    "Cricket pitch",
    "Cricket ball",
    "Cricket bat",
    "Dismissal (cricket)",
    "Innings",
    "Over (cricket)",
    "Wide (cricket)",
    "No-ball",
    "LBW",
    "Run out",
    "Stumping",

    # Formats
    "Test cricket",
    "One Day International",
    "Twenty20",
    "The Hundred (cricket competition)",

    # Batting & Bowling
    "Batting (cricket)",
    "Bowling (cricket)",
    "Spin bowling",
    "Fast bowling",
    "Swing bowling",
    "Century (cricket)",
    "Duck (cricket)",

    # Famous Players
    "Sachin Tendulkar",
    "Virat Kohli",
    "Brian Lara",
    "Shane Warne",
    "Ricky Ponting",
    "MS Dhoni",
    "Wasim Akram",
    "Imran Khan",
    "Viv Richards",
    "Don Bradman",

    # Tournaments
    "ICC Cricket World Cup",
    "Indian Premier League",
    "The Ashes",
    "ICC World Twenty20",
    "Champions Trophy",
    "Indian Premier League records",

    # Teams
    "Indian cricket team",
    "Australian cricket team",
    "England cricket team",
    "Pakistan cricket team",
    "West Indies cricket team",
]

# Save folder
os.makedirs("cricket_docs", exist_ok=True)

for topic in cricket_topics:
    page = wiki.page(topic)
    
    if page.exists():
        filename = topic.replace(" ", "_").replace("/", "_") + ".txt"
        filepath = os.path.join("cricket_docs", filename)
        
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(page.text)
        
        print(f"✅ Saved: {filename} ({len(page.text)} chars)")
    else:
        print(f"❌ Not found: {topic}")
```
